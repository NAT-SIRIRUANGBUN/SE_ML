```python
import sys
sys.path.append("./lib")

from lib.whisper_handmade import Whisper
import pandas as pd
```

```python
model = Whisper(model='large')
```

```python
df = pd.read_csv('./Categorized-Data/Gowajee-Corpus/thai_foreign/final_data.csv')
```

```python
ans_list = []
```

```python
for i in range(df.shape[0]):
    try:
        ans = model.run_full(f'./Categorized-Data/Gowajee-Corpus/thai_foreign/audio/{df.iloc[i]['audio']}')
    except:
        ans = "ERROR"
    print(f'{i}: {ans}')
    ans_list.append(ans)
    
```

```python
df = pd.DataFrame(ans_list)
```

```python
df.to_csv('Gowajee-Thai-Foreign-Whisper-Large.csv', index=False)
```
